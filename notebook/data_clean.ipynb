```python
kaggle_df = kaggle_df.rename(columns={'Strain':'name', 'Type':'race'})
```

```python
merged = df.merge(kaggle_df, how='inner', on=['name', 'race'])
```

```python
merged
```

```python
# combine flavors:
for index, row in merged.iterrows():
    for flav in row['Flavor'].split(','):
        if flav not in row['flavors']:
            row['flavors'].append(flav)

merged = merged.drop('Flavor', axis=1)
```

```python
# combine positive effects:
for index, row in merged.iterrows():
    for effect in row['Effects'].split(','):
        if effect not in row['positive']:
            row['positive'].append(effect)
            
merged = merged.drop('Effects', axis=1)
```

```python

merged = merged.set_index('id')
```

```python
merged
```

```python
merged['positive']=merged['positive'].apply(', '.join)
```

```python
merged['negative'] = merged['negative'].apply(', '.join) 
```

```python
merged['medical'] = merged['medical'].apply(', '.join) 
```

```python
merged['flavors'] = merged['flavors'].apply(', '.join) 
```
